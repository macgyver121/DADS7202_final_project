```python
from transformers import AutoTokenizer, AutoModelWithLMHead

tokenizer = AutoTokenizer.from_pretrained("gagan3012/k2t")

model = AutoModelWithLMHead.from_pretrained("gagan3012/k2t")

# Encode the keyword as a sequence of tokens
keywords = "India food expensive 20 50 60"
input_ids = tokenizer.encode(keywords, return_tensors='pt')

# Use the model to generate a sequence of tokens
output = model.generate(input_ids, max_length=50, min_length=50)

# Decode the output tokens to text
output_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(output_text)
```

```python
import numpy as np
from transformers import AutoTokenizer

data = 'hello'

tokenizer = AutoTokenizer.from_pretrained("gagan3012/k2t")
tokenized_data = tokenizer(data, return_tensors="np", padding=True)
```

```python
print(tokenized_data)
```

```python
labels = 'hello world my name is mac'
```

```python
from transformers import TFAutoModelForSequenceClassification
from tensorflow.keras.optimizers import Adam

# Load and compile our model
model = AutoModelWithLMHead.from_pretrained("gagan3012/k2t")
# Lower learning rates are often better for fine-tuning transformers
#model.compile(optimizer=Adam(3e-5))

model.fit(tokenized_data, labels)
```

```python
lst = ['mac', 'hello', 'world']
for i in lst : 
    input_ids = tokenizer.encode(i, return_tensors='pt')
    labels = tokenizer.encode('hello world my name is mac and you are someone that very important', return_tensors='pt')
    # the forward function automatically creates the correct decoder_input_ids
    loss = model(input_ids=input_ids, labels=labels).loss
    print(loss)
```

```python
for i in range(0,1000) :
    input_ids = tokenizer("The house is wonderful", return_tensors="pt").input_ids
    labels = tokenizer("The house is wonderful and have a great view. This is tha place like heaven The house is wonderful and have a great view. This is tha place like heaven", return_tensors="pt").input_ids

    # the forward function automatically creates the correct decoder_input_ids
    loss = model(input_ids=input_ids, labels=labels).loss
    loss.item()
```
